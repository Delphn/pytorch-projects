```python
import torch
from torch.nn.functional import normalize

t = torch.tensor([[1.,2.,3.],[4.,5.,6.]])
print("Tensor:", t)
```

```python
t1 = normalize(t, p=1.0, dim = 1)
t2 = normalize(t, p=2.0, dim = 0)

print("Normalized tensor:\n", t1)
print("Normalized tensor:\n", t2)
```

```python
import requests
import io

image = requests.get('https://images.unsplash.com/photo-1506328177866-9e307a27fbca?crop=entropy&cs=tinysrgb&fit=max&fm=jpg&ixid=MnwxMTc3M3wwfDF8c2VhcmNofDQzfHxhbmltYWwlMjBqdW1wfGVufDB8fHx8MTYyMzAxNjgyOQ&ixlib=rb-1.2.1&q=80&w=2000')
image_file = Image.open(io.BytesIO(image.content))
image_file
```

```python
# Pixels distribution
import matplotlib.pyplot as plt
import numpy as np

plt.hist(np.array(image_file).ravel(), bins=50, density=True)
plt.xlabel('Pixel value')
plt.ylabel('relative frequency')
plt.title('Pixels distribution')
```

```python
# normalize
import torchvision

transform = torchvision.transforms.Compose([
  torchvision.transforms.ToTensor(),
  torchvision.transforms.Normalize(
    mean=[0.485, 0.456, 0.406],
    std=[0.229, 0.224, 0.225]
  ),
])

normalized_img = transform(image_file)
```

```python
plt.hist(normalized_img.numpy().ravel(), bins=30, density=True)
plt.xlabel('Pixel values')
plt.ylabel('relative frequency')
plt.title('Normalized pixels distribution')
```
